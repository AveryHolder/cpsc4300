```python
'''Upload another csv into a table'''
dog_show_breed_df = pd.read_csv('Dog_Data_KIB(BESTINSHOW).csv')
dog_show_breed_df
```

```python
'''Rename Dog breed to Breed to be consistent'''
dog_show_breed_df = dog_show_breed_df.rename(columns={'Dog breed':'Breed', 'LIFETIME COST, $': 'Lifetime Cost', 'Intelligence, Percentage': 'Intelligence %'})
dog_show_breed_df
```

```python
'''Need to change breed names to add a - so we can compare data tables'''
```

```python
'''Since both of these dataframes have different breeds, we will have to do some data clipping.'''
double_dogs = []
for breed in dog_show_breed_df['Breed']:
    breed = str(breed).replace(' ', '-')
    if breed in list(dog_df['Breed']):
        double_dogs.append(breed)
double_dogs
```

```python
'''Create a dataframe with all of the information from both datasets for double dogs'''
dd_attr = dog_df.columns.append(dog_show_breed_df.columns)
double_df_data = dog_show_breed_df.set_index('Breed').combine_first(dog_df.drop_duplicates().set_index('Breed'))
double_dogs_df = pd.DataFrame(index=double_dogs, columns=dd_attr, data=double_df_data)
double_dogs_df = double_dogs_df.drop(columns={'Breed', 'Category2'})
```

```python
'''Scale size into 0,1,2'''
double_dogs_df = double_dogs_df.replace({'small': 0, 'medium': 1, 'large': 2, 'no data': 'NaN'})
double_dogs_df = double_dogs_df.dropna()
```

```python
double_dogs_df.sort_values(by="Lifetime Cost", ascending=False)
```

```python
'''Here, we will derive attributes to create ratio values in the dataframe'''
attrs = ['Trainability', 'All Around Friendliness', 'Health And Grooming Needs', 'Size']
derived_attrs = []
for attr1 in attrs[:-1]:
    for attr2 in attrs[:-1]:
        if attr2 == attr1:
            continue
        derived_attr = attr1+'_'+attr2
        derived_attrs.append(derived_attr)
        dog_df[derived_attr] = dog_df[attr1]/dog_df[attr2]
```

```python
''''''
def corr_with(attribute):
    corrs = dog_df.corr()[attribute].drop(index=attribute).sort_values(ascending=False)
    _ = corrs.plot.barh(title='Correlation Coefficient b/t attribute with ' + attribute)
    print(corrs)
```

```python
dog_df = dog_df.set_index('Breed')
```

```python
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns

train_data, test_data = train_test_split(dog_df, test_size=0.33, random_state=6300)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6))
for idx, col in enumerate(['Trainability', 'Mean Score']):
    for df in [dog_df, train_data, test_data]:
        _ = sns.distplot(df[col], bins=20, ax=axes[idx])
    _ = axes[idx].set_xlim(dog_df[col].min(), dog_df[col].max())
```

```python
train_data.head()
```

```python
X_train, y_train = train_data[train_data.columns[:-1]], train_data[train_data.columns[-1]]
X_test, y_test = test_data[test_data.columns[:-1]], test_data[test_data.columns[-1]]
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
base_model = reg.fit(X_train, y_train)
```

```python
from collections  import defaultdict
from sklearn.metrics import mean_squared_error
from sklearn.metrics import explained_variance_score
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score

model_perf = defaultdict(dict)
def compute_model_perf(model_name, tag, y_true, y_pred):
    model_perf[model_name]['{}.explained_variance'.format(tag)] = explained_variance_score(y_true, y_pred)
    model_perf[model_name]['{}.RSME'.format(tag)] = mean_squared_error(y_true, y_pred, squared=False)
    model_perf[model_name]['{}.R2'.format(tag)] = r2_score(y_true, y_pred)
    model_perf[model_name]['{}.RelativeError'.format(tag)] = mean_absolute_error(y_true/y_true, y_pred/y_true)
```

```python
y_train_pred = base_model.predict(X_train)
y_test_pred = base_model.predict(X_test)
compute_model_perf('LinearRegressionBase', 'Train', y_train, y_train_pred)
compute_model_perf('LinearRegressionBase', 'Test', y_test, y_test_pred)
```

```python
df_perf = pd.DataFrame(model_perf).T
df_perf
```

```python
fig, ax = plt.subplots(1, 2, figsize=(14, 6))
_ = sns.scatterplot(x=y_train, y=(y_train-y_train_pred), ax = ax[0], label='Train')
_ = sns.scatterplot(x=y_test, y=(y_test-y_test_pred), ax = ax[0], label='Test')
ax[0].set_ylabel('Absolute Error')
_ = sns.scatterplot(x=y_train, y=(1.0-y_train_pred/y_train), ax = ax[1], label='Train')
_ = sns.scatterplot(x=y_test, y=(1.0-y_test_pred/y_test), ax = ax[1], label='Test')
_ = ax[1].set_ylabel('Relative Error')
```

```python
from sklearn.neighbors import KNeighborsRegressor

model_perf = defaultdict(dict)

y = train_data[train_data.columns[-1]]
y_test = test_data[test_data.columns[-1]]

features = train_data.columns[:-1]
X = train_data[features]
X_test = test_data[features]

for n_neighbors in range(1, 11):
    
    neigh = KNeighborsRegressor(n_neighbors)
    neigh.fit(X, y)
    
    y_train_pred = neigh.predict(X)
    y_test_pred = neigh.predict(X_test)
    
    model_name =  "KNeighborsRegressor-{}".format(n_neighbors)
    compute_model_perf(model_name, 'Train', y, y_train_pred)
    compute_model_perf(model_name, 'Test', y_test, y_test_pred)

    
neigh.fit(X_test, y_test)
```

```python
df_perf = pd.DataFrame(model_perf).T
df_perf
```
